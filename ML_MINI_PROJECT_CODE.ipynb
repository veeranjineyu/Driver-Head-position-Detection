```python
#to solve the problem of class imbalance doing the OVERsampling. It increases the likelihood of overfitting since it replicates the minority class events.

import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import seaborn as sns
X = pd.read_csv("/content/drivePoints.csv") # reading the Dataset

X = X.sample(frac = 1) # suffling


#dropping the useless features.

X = X.drop(['fileName','subject','imgNum','ang'],axis = 1)

y = X['label']


class_count_2, class_count_3,class_count_1 = y.value_counts()

# Separate class
class_1 = X[y == 1]
class_2 = X[y == 2]# print the shape of the class
class_3 = X[y == 3]
print('class 1:', class_1.shape)
print('class 2:', class_2.shape)
print('class 3:', class_3.shape)

class_1_over = class_1.sample(class_count_2, replace=True)
class_3_over = class_3.sample(class_count_2, replace=True)

X = pd.concat([class_2, class_1_over,class_3_over], axis=0)

class_1 = X[y == 1]
class_2 = X[y == 2]# print the shape of the class
class_3 = X[y == 3]
print('class 1:', class_1.shape)
print('class 2:', class_2.shape)
print('class 3:', class_3.shape)
class1 = 0
class2 = 0
class3 = 0


df = X
y = X['label']
y = np.array(y)

for i in range(len(y)) : 
  if(y[i]==1) : 
    class1+=1
  if (y[i] == 2) : 
    class2+=1
  if (y[i] == 3) : 
    class3+=1
        


#print("y=",y)
X = X.drop(['label'],axis=1)

X = np.array(X)

#print(df) 

X = X.astype(float)

# doing the normalization 
max_value = np.max(X)

for i in range(len(X)) : 
  for j in range(len(X[0])) : 
    X[i][j] = X[i][j]/max_value

print(X.shape)    

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=0)



# training and testing the model by using logistic regression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)

#print("y_pred = ",y_pred)
print("AFTER DOING THE OVERSAMPLING :")
print("accuracy by using logistic regression=",logreg.score(X_test,y_test))

from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
print("cnf_matrix=",cnf_matrix)

class_accuracy_one = 0
sum = 0
for i in range(len(cnf_matrix)) : 
  sum+=cnf_matrix[i][0]
class_wise_accuracy_one = cnf_matrix[0][0]/sum

print("class_wise_accuracy_one=",class_wise_accuracy_one)

class_accuracy_two = 0
sum = 0
for i in range(len(cnf_matrix)) : 
  sum+=cnf_matrix[i][1]
class_wise_accuracy_two = cnf_matrix[1][1]/sum

print("class_wise_accuracy_two=",class_wise_accuracy_two)

class_accuracy_three = 0
sum = 0
for i in range(len(cnf_matrix)) : 
  sum+=cnf_matrix[i][2]
class_wise_accuracy_three = cnf_matrix[2][2]/sum

print("class_wise_accuracy_three=",class_wise_accuracy_three)



classes = list([1,2,3])
values = list([class_wise_accuracy_one,class_wise_accuracy_two,class_wise_accuracy_three])
print("values=",values)

fig = plt.figure(figsize = (10, 5))

plt.bar(classes, values, color ='green',width = 0.4)

plt.xlabel("classes")
plt.ylabel("acuuracy")
plt.title("accuracy of different classes using oversampling")
plt.show()
```
